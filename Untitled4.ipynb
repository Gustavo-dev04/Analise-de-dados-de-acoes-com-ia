```python
# 📌 IMPORTAÇÕES ESSENCIAIS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime, timedelta
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 📌 CONFIGURAÇÕES
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (18, 9)
SYMBOL = 'NVDA'
DAYS_TO_FETCH = 365 * 3  # 3 anos de dados
TEST_SIZE = 0.2
LEARNING_RATE = 0.01
EPOCHS = 1000

# ======================
# 📌 1. BAIXAR E PROCESSAR DADOS
# ======================
def fetch_and_preprocess_data():
    """Obtém e prepara os dados para modelagem"""
    try:
        end_date = datetime.now()
        start_date = end_date - timedelta(days=DAYS_TO_FETCH)

        # Busca dados históricos
        df = yf.download(SYMBOL, start=start_date, end=end_date, progress=False)
        if df.empty:
            raise ValueError("Nenhum dado retornado do Yahoo Finance")

        # Seleciona e renomeia colunas
        df = df[['Close', 'Volume']].copy()
        df.columns = ['close', 'volume']

        # Feature engineering
        df['return'] = df['close'].pct_change()

        # Médias móveis
        for window in [5, 10, 20, 50]:
            df[f'ma_{window}'] = df['close'].rolling(window, min_periods=1).mean()
            df[f'volatility_{window}'] = df['return'].rolling(window, min_periods=1).std()

        # Tendência
        df['trend'] = (df['close'] > df['ma_20']).astype(int)

        # Remove linhas com NaN
        df.dropna(inplace=True)

        return df

    except Exception as e:
        raise Exception(f"Erro no pré-processamento: {str(e)}")

# ======================
# 📌 2. REGRESSÃO LINEAR COM GRADIENTE DESCENDENTE
# ======================
class LinearRegressionGD:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
        self.scaler = RobustScaler()
        self.feature_cols = [
            'close', 'volume', 'ma_5', 'ma_20',
            'volatility_20', 'trend'
        ]

    def prepare_data(self, df):
        """Prepara os dados para treinamento"""
        try:
            # Verifica colunas necessárias
            missing_cols = [col for col in self.feature_cols if col not in df.columns]
            if missing_cols:
                raise ValueError(f"Colunas faltando: {missing_cols}")

            # Cria X e y
            X = df[self.feature_cols].copy()
            y = df['close'].shift(-1).dropna().copy()

            # Alinha os índices
            X = X.iloc[:-1]
            X, y = X.align(y, axis=0, copy=False)

            # Normalização
            X_scaled = self.scaler.fit_transform(X)

            # Divisão treino-teste temporal
            split_idx = int(len(X) * (1 - TEST_SIZE))
            X_train, X_test = X_scaled[:split_idx], X_scaled[split_idx:]
            y_train, y_test = y[:split_idx], y[split_idx:]

            # Adiciona coluna de bias (intercept)
            X_train = np.c_[np.ones(X_train.shape[0]), X_train]
            X_test = np.c_[np.ones(X_test.shape[0]), X_test]

            return X_train, X_test, y_train, y_test

        except Exception as e:
            raise Exception(f"Erro na preparação dos dados: {str(e)}")

    def train(self, X, y):
        """Treina o modelo usando gradiente descendente"""
        try:
            n_samples, n_features = X.shape
            self.weights = np.zeros(n_features)
            self.bias = 0
            cost_history = []

            for epoch in range(self.epochs):
                # Previsões
                y_pred = np.dot(X, self.weights) + self.bias

                # Gradientes
                dw = (1/n_samples) * np.dot(X.T, (y_pred - y))
                db = (1/n_samples) * np.sum(y_pred - y)

                # Atualização dos parâmetros
                self.weights -= self.learning_rate * dw
                self.bias -= self.learning_rate * db

                # Cálculo do custo (MSE)
                cost = np.mean((y_pred - y)**2)
                cost_history.append(cost)

                # Print progresso a cada 100 épocas
                if epoch % 100 == 0:
                    print(f"Época {epoch}: Custo = {cost:.4f}")

            return cost_history

        except Exception as e:
            raise Exception(f"Erro no treinamento: {str(e)}")

    def predict(self, X):
        """Faz previsões usando os parâmetros aprendidos"""
        try:
            return np.dot(X, self.weights) + self.bias

        except Exception as e:
            raise Exception(f"Erro na previsão: {str(e)}")

    def evaluate(self, X_test, y_test):
        """Avalia o modelo e plota resultados"""
        try:
            y_pred = self.predict(X_test)

            # Garante alinhamento
            y_test, y_pred = y_test.align(pd.Series(y_pred, index=y_test.index), axis=0)

            # Cálculo de métricas
            mae = mean_absolute_error(y_test, y_pred)
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            r2 = r2_score(y_test, y_pred)

            # Plot
            plt.figure(figsize=(18, 9))
            plt.plot(y_test.values, label='Valor Real', color='#1f77b4', linewidth=2)
            plt.plot(y_pred.values, label='Previsão', color='#ff7f0e', linestyle='--')

            # Destaque para grandes erros
            errors = np.abs(y_test.values - y_pred.values)
            large_errors = errors > np.percentile(errors, 90)
            plt.scatter(
                np.where(large_errors)[0],
                y_test.values[large_errors],
                color='red',
                s=100,
                label='Grandes discrepâncias'
            )

            plt.title(f'Previsão com Regressão Linear (MAE: {mae:.2f}, R²: {r2:.2f})', fontsize=16)
            plt.xlabel('Dias', fontsize=12)
            plt.ylabel('Preço ($)', fontsize=12)
            plt.legend()
            plt.grid(alpha=0.2)
            plt.show()

            return {'mae': mae, 'rmse': rmse, 'r2': r2}

        except Exception as e:
            raise Exception(f"Erro na avaliação: {str(e)}")

# ======================
# 📌 3. EXECUÇÃO PRINCIPAL
# ======================
if __name__ == "__main__":
    print("=== SISTEMA DE PREVISÃO COM REGRESSÃO LINEAR E GRADIENTE DESCENDENTE ===")

    try:
        # 1. Obtenção de dados
        print("🔄 Obtendo e processando dados...")
        df = fetch_and_preprocess_data()
        print(f"✅ Dados carregados. Período: {df.index[0].date()} a {df.index[-1].date()}")

        # 2. Preparação do modelo
        print("\n🧠 Preparando modelo...")
        model = LinearRegressionGD(learning_rate=LEARNING_RATE, epochs=EPOCHS)
        X_train, X_test, y_train, y_test = model.prepare_data(df)
        print(f"📚 Dados divididos - Treino: {len(X_train)}, Teste: {len(X_test)}")

        # 3. Treinamento
        print("\n🔮 Treinando modelo com gradiente descendente...")
        cost_history = model.train(X_train, y_train)

        # Plot da função de custo
        plt.figure(figsize=(12, 6))
        plt.plot(cost_history)
        plt.title('Convergência do Gradiente Descendente')
        plt.xlabel('Época')
        plt.ylabel('Custo (MSE)')
        plt.show()

        # 4. Avaliação
        print("\n📊 Avaliando desempenho...")
        metrics = model.evaluate(X_test, y_test)

        print("\n🔍 Métricas Finais:")
        print(f"- MAE (Erro Médio Absoluto): ${metrics['mae']:.2f}")
        print(f"- RMSE (Erro Quadrático Médio): ${metrics['rmse']:.2f}")
        print(f"- R² (Variância Explicada): {metrics['r2']:.2%}")

    except Exception as e:
        print(f"\n❌ Erro no processo: {str(e)}")
        print("\nSoluções possíveis:")
        print("- Verifique sua conexão com a internet")
        print("- Tente aumentar DAYS_TO_FETCH para obter mais dados")
        print("- Verifique se o símbolo da ação está correto")
```
