```python
# üìå IMPORTA√á√ïES ESSENCIAIS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime, timedelta
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# üìå CONFIGURA√á√ïES
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (18, 9)
SYMBOL = 'NVDA'
DAYS_TO_FETCH = 365 * 5  # 5 anos de dados
TEST_SIZE = 0.2
LEARNING_RATE = 0.05
EPOCHS = 300000 # Reduzido para demonstra√ß√£o

# ======================
# üìå 1. BAIXAR E PROCESSAR DADOS
# ======================
def fetch_and_preprocess_data():
    """Obt√©m e prepara os dados para modelagem"""
    try:
        end_date = datetime.now()
        start_date = end_date - timedelta(days=DAYS_TO_FETCH)

        # Busca dados hist√≥ricos
        df = yf.download(SYMBOL, start=start_date, end=end_date, progress=False)
        if df.empty:
            raise ValueError("Nenhum dado retornado do Yahoo Finance")

        # Seleciona e renomeia colunas
        df = df[['Close', 'Volume']].copy()
        df.columns = ['close', 'volume']

        # Feature engineering
        df['return'] = df['close'].pct_change()

        # M√©dias m√≥veis
        for window in [5, 10, 20, 50]:
            df[f'ma_{window}'] = df['close'].rolling(window, min_periods=1).mean()
            df[f'volatility_{window}'] = df['return'].rolling(window, min_periods=1).std()

        # Tend√™ncia
        df['trend'] = (df['close'] > df['ma_20']).astype(int)

        # Remove linhas com NaN
        df.dropna(inplace=True)

        return df

    except Exception as e:
        raise Exception(f"Erro no pr√©-processamento: {str(e)}")

# ======================
# üìå 2. REGRESS√ÉO LINEAR COM GRADIENTE DESCENDENTE
# ======================
class LinearRegressionGD:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
        self.scaler = RobustScaler()
        self.feature_cols = [
            'close', 'volume', 'ma_5', 'ma_20',
            'volatility_20', 'trend'
        ]

    def prepare_data(self, df):
        """Prepara os dados para treinamento"""
        try:
            # Verifica colunas necess√°rias
            missing_cols = [col for col in self.feature_cols if col not in df.columns]
            if missing_cols:
                raise ValueError(f"Colunas faltando: {missing_cols}")

            # Cria X e y
            X = df[self.feature_cols].copy()
            y = df['close'].shift(-1).dropna().copy()

            # Alinha os √≠ndices
            X = X.iloc[:-1]
            X, y = X.align(y, axis=0, copy=False)

            # Normaliza√ß√£o
            X_scaled = self.scaler.fit_transform(X)

            # Divis√£o treino-teste temporal
            split_idx = int(len(X) * (1 - TEST_SIZE))
            X_train, X_test = X_scaled[:split_idx], X_scaled[split_idx:]
            y_train, y_test = y[:split_idx], y[split_idx:]

            # Adiciona coluna de bias (intercept)
            X_train = np.c_[np.ones(X_train.shape[0]), X_train]
            X_test = np.c_[np.ones(X_test.shape[0]), X_test]

            return X_train, X_test, y_train, y_test

        except Exception as e:
            raise Exception(f"Erro na prepara√ß√£o dos dados: {str(e)}")

    def train(self, X, y):
        """Treina o modelo usando gradiente descendente"""
        try:
            n_samples, n_features = X.shape
            self.weights = np.zeros(n_features)
            self.bias = 0
            cost_history = []

            for epoch in range(self.epochs):
                # Previs√µes
                y_pred = np.dot(X, self.weights) + self.bias

                # Gradientes
                dw = (1/n_samples) * np.dot(X.T, (y_pred - y))
                db = (1/n_samples) * np.sum(y_pred - y)

                # Atualiza√ß√£o dos par√¢metros
                self.weights -= self.learning_rate * dw
                self.bias -= self.learning_rate * db

                # C√°lculo do custo (MSE)
                cost = np.mean((y_pred - y)**2)
                cost_history.append(cost)

                # Print progresso a cada 10% das √©pocas
                if epoch % (self.epochs//10) == 0:
                    print(f"√âpoca {epoch}: Custo = {cost:.4f}")

            return cost_history

        except Exception as e:
            raise Exception(f"Erro no treinamento: {str(e)}")

    def predict(self, X):
        """Faz previs√µes usando os par√¢metros aprendidos"""
        try:
            return np.dot(X, self.weights) + self.bias

        except Exception as e:
            raise Exception(f"Erro na previs√£o: {str(e)}")

    def evaluate(self, X_test, y_test):
        """Avalia o modelo e plota resultados"""
        try:
            y_pred = self.predict(X_test)

            # Garante alinhamento
            y_test, y_pred = y_test.align(pd.Series(y_pred, index=y_test.index), axis=0)

            # C√°lculo de m√©tricas
            mae = mean_absolute_error(y_test, y_pred)
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            r2 = r2_score(y_test, y_pred)

            # Plot
            plt.figure(figsize=(18, 9))
            plt.plot(y_test.values, label='Valor Real', color='#1f77b4', linewidth=2)
            plt.plot(y_pred.values, label='Previs√£o', color='#ff7f0e', linestyle='--')

            # Destaque para grandes erros
            errors = np.abs(y_test.values - y_pred.values)
            large_errors = errors > np.percentile(errors, 90)
            plt.scatter(
                np.where(large_errors)[0],
                y_test.values[large_errors],
                color='red',
                s=100,
                label='Grandes discrep√¢ncias'
            )

            plt.title(f'Previs√£o com Regress√£o Linear (MAE: {mae:.2f}, R¬≤: {r2:.2f})', fontsize=16)
            plt.xlabel('Dias', fontsize=12)
            plt.ylabel('Pre√ßo ($)', fontsize=12)
            plt.legend()
            plt.grid(alpha=0.2)
            plt.show()

            return {'mae': mae, 'rmse': rmse, 'r2': r2}

        except Exception as e:
            raise Exception(f"Erro na avalia√ß√£o: {str(e)}")

    def forecast(self, df, days=7):
        """
        Preve os pr√≥ximos 'days' dias com base nos dados mais recentes

        Args:
            df: DataFrame com os dados hist√≥ricos
            days: N√∫mero de dias para prever (padr√£o=7)

        Returns:
            DataFrame com as previs√µes e datas futuras
        """
        try:
            # Cria c√≥pia para n√£o modificar o original
            df_temp = df.copy()

            # Lista para armazenar previs√µes
            forecasts = []
            dates = []

            # √öltima data dispon√≠vel
            last_date = df_temp.index[-1]

            # Prepara os dados mais recentes para previs√£o
            recent_data = df_temp[self.feature_cols].iloc[-1:].copy()

            for i in range(1, days+1):
                # Normaliza os dados
                data_scaled = self.scaler.transform(recent_data)

                # Adiciona coluna de bias
                data_scaled = np.c_[np.ones(data_scaled.shape[0]), data_scaled]

                # Faz a previs√£o
                predicted_price = self.predict(data_scaled)[0]
                forecasts.append(predicted_price)

                # Data da previs√£o
                new_date = last_date + timedelta(days=i)
                dates.append(new_date)

                # Atualiza os dados para a pr√≥xima previs√£o
                new_row = recent_data.iloc[-1:].copy()

                # Atualiza o pre√ßo com a previs√£o
                new_row['close'] = predicted_price

                # Atualiza m√©dias m√≥veis (simplificado)
                for window in [5, 10, 20, 50]:
                    if f'ma_{window}' in new_row.columns:
                        new_row[f'ma_{window}'] = predicted_price

                # Atualiza tend√™ncia (assume que continua a mesma)
                if 'trend' in new_row.columns:
                    new_row['trend'] = 1 if predicted_price > new_row['ma_20'].values[0] else 0

                # Adiciona ao DataFrame tempor√°rio para c√°lculos futuros
                recent_data = pd.concat([recent_data, new_row])

                # Mant√©m apenas a √∫ltima linha para a pr√≥xima itera√ß√£o
                recent_data = recent_data.iloc[-1:]

            # Cria DataFrame com resultados
            df_forecast = pd.DataFrame({
                'date': dates,
                'forecast': forecasts
            }).set_index('date')

            return df_forecast

        except Exception as e:
            raise Exception(f"Erro na previs√£o: {str(e)}")

# ======================
# üìå 3. EXECU√á√ÉO PRINCIPAL
# ======================
if __name__ == "__main__":
    print("=== SISTEMA DE PREVIS√ÉO COM REGRESS√ÉO LINEAR E GRADIENTE DESCENDENTE ===")

    try:
        # 1. Obten√ß√£o de dados
        print("üîÑ Obtendo e processando dados...")
        df = fetch_and_preprocess_data()
        print(f"‚úÖ Dados carregados. Per√≠odo: {df.index[0].date()} a {df.index[-1].date()}")
        print(f"üìä N√∫mero de registros: {len(df)}")
        print(f"üìà √öltimo pre√ßo: ${df['close'].iloc[-1]:.2f}")

        # 2. Prepara√ß√£o do modelo
        print("\nüß† Preparando modelo...")
        model = LinearRegressionGD(learning_rate=LEARNING_RATE, epochs=EPOCHS)
        X_train, X_test, y_train, y_test = model.prepare_data(df)
        print(f"üìö Dados divididos - Treino: {len(X_train)}, Teste: {len(X_test)}")

        # 3. Treinamento
        print("\nüîÆ Treinando modelo com gradiente descendente...")
        cost_history = model.train(X_train, y_train)

        # Plot da fun√ß√£o de custo
        plt.figure(figsize=(12, 6))
        plt.plot(cost_history)
        plt.title('Converg√™ncia do Gradiente Descendente')
        plt.xlabel('√âpoca')
        plt.ylabel('Custo (MSE)')
        plt.show()

        # 4. Avalia√ß√£o
        print("\nüìä Avaliando desempenho...")
        metrics = model.evaluate(X_test, y_test)

        print("\nüîç M√©tricas Finais:")
        print(f"- MAE (Erro M√©dio Absoluto): ${metrics['mae']:.2f}")
        print(f"- RMSE (Erro Quadr√°tico M√©dio): ${metrics['rmse']:.2f}")
        print(f"- R¬≤ (Vari√¢ncia Explicada): {metrics['r2']:.2%}")

        # 5. Previs√£o para os pr√≥ximos 7 dias
        print("\nüîÆ Fazendo previs√£o para os pr√≥ximos 7 dias...")
        df_forecast = model.forecast(df)

        print("\nüìà Previs√£o para os pr√≥ximos 7 dias:")
        print(df_forecast.round(2))

        # Plot das previs√µes
        plt.figure(figsize=(18, 9))

        # Plot dos √∫ltimos 30 dias hist√≥ricos
        plt.plot(df['close'].iloc[-30:], label='Hist√≥rico (30 dias)',
                color='#1f77b4', marker='o')

        # Plot das previs√µes
        plt.plot(df_forecast['forecast'], label='Previs√£o (7 dias)',
                color='red', linestyle='--', marker='o')

        plt.title(f'Previs√£o de Pre√ßos para {SYMBOL} - Pr√≥ximos 7 Dias', fontsize=16)
        plt.xlabel('Data', fontsize=12)
        plt.ylabel('Pre√ßo ($)', fontsize=12)
        plt.legend()
        plt.grid(alpha=0.2)
        plt.show()

    except Exception as e:
        print(f"\n‚ùå Erro no processo: {str(e)}")
        print("\nSolu√ß√µes poss√≠veis:")
        print("- Verifique sua conex√£o com a internet")
        print("- Tente aumentar DAYS_TO_FETCH para obter mais dados")
        print("- Verifique se o s√≠mbolo da a√ß√£o est√° correto")
        print("- Reduza o n√∫mero de √©pocas (EPOCHS) se o treinamento estiver demorando muito")
```
